# Great Figures

#### Publication Quality Plots in Python using Matplotlib

For those of you adventurous enough to jump into the Python boat even for your scientific computing needs, here are some useful settings for Matplotlib that you might want to set in your `~/.matplotlib/matplotlibrc` file. For a complete listing of all the options and their defaults, please refer to your installation's site-packages/matplotlib/mpl-data/matplotlibrc file or check out the matplotlib customization page. Note that most, if not all, these options can be directly modified from a python script or the console as explained here. 

Computer Modern Fonts - to match the default fonts in LaTeX? 
* font.family : serif 
* font.serif : cmr10 
* text.usetex : True 

Plot Sizing - balancing good rendering on-screen with low amount of whitespace around plot 
* figure.figsize : 10, 8 
* figure.subplot.left : 0.08 
* figure.subplot.right : 0.92 
* figure.subplot.bottom : 0.08 
* figure.subplot.top : 0.92 
* figure.subplot.wspace : 0.25 
* figure.subplot.hspace : 0.25 

Some IEEE Reqs - figures should at least be 300dpi and fonts must be Truetype 
* savefig.dpi : 300 
* ps.fonttype : 42 
* pdf.fonttype : 42

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

rcParams['font.family']= 'serif'
rcParams['font.weight']= 'bold'
rcParams['font.serif']= 'cmr10'
rcParams['font.size']= 22
rcParams['text.usetex'] = True
rcParams['figure.figsize'] = 10, 8
rcParams['figure.subplot.left'] = 0.08 
rcParams['figure.subplot.right'] = 0.92 
rcParams['figure.subplot.bottom'] = 0.08 
rcParams['figure.subplot.top'] = 0.92 
rcParams['figure.subplot.wspace'] = 0.25 
rcParams['figure.subplot.hspace'] = 0.25 
rcParams['savefig.dpi'] = 300 
rcParams['ps.fonttype'] = 42 
rcParams['pdf.fonttype'] = 42
```
